# Scraping Smartphones dataset

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# getting html code of smartphones data (acquired using selenium) 
with open("C:\\Users\\Lenovo\\Desktop\\smartphone.html","r", encoding= "utf-8") as file_object:
    html_code = file_object.read()
```

```python
soup = BeautifulSoup(html_code,"lxml")
```

```python
containers = soup.find_all("div",{"class":"sm-product has-tag has-features has-actions"})
```

```python
names= []
price= []
ratings= []
sims= []
processor= []
ram= []
battery= []
display= []
camera= []
card= []
os= []

for i in containers:
    try:
        names.append(i.find("h2").text)
    except:
        names.append(np.nan)

    try:
        price.append(i.find("span",{"class":"price"}).text)
    except:
        price.append(np.nan)

    score1 = i.find("div",{"class":"score rank-1-bg"})
    score2 = i.find("div",{"class":"score rank-2-bg"})
    score3 = i.find("div",{"class":"score rank-3-bg"})
    
    if score1 and score1.find("b"):
        ratings.append(i.find("div",{"class":"score rank-1-bg"}).find("b").text)
    elif score2 and score2.find("b"):
        ratings.append(i.find("div",{"class":"score rank-2-bg"}).find("b").text)
    elif score3 and score3.find("b"):
        ratings.append(i.find("div",{"class":"score rank-3-bg"}).find("b").text)
    else:
        ratings.append(np.nan)

    ul_tag = i.find("ul", {"class": "sm-feat specs"})
    x = ul_tag.find_all("li") if ul_tag else []

    try:
        sims.append(x[0].text) # sims
    except:
        sims.append(np.nan)

    try:
        processor.append(x[1].text) # processor
    except:
        processor.append(np.nan)

    try:
        ram.append(x[2].text) # ram
    except:
        ram.append(np.nan)

    try:
        battery.append(x[3].text) # battery
    except:
        battery.append(np.nan)

    try:    
        display.append(x[4].text) # display
    except:
        display.append(np.nan)

    try:
        camera.append(x[5].text) # camera
    except:
        camera.append(np.nan)

    try:
        card.append(x[6].text) # card
    except:
        card.append(np.nan)

    try:
        os.append(x[7].text) # os
    except:
        os.append(np.nan)
    
    
```

```python
print("names     :", len(names))
print("price     :", len(price))
print("ratings   :", len(ratings))
print("sims      :", len(sims))
print("processor :", len(processor))
print("ram       :", len(ram))
print("battery   :", len(battery))
print("display   :", len(display))
print("camera    :", len(camera))
print("card      :", len(card))
print("os        :", len(os))
```

```python
df = pd.DataFrame({
    'model': names,
    'price': price,
    'rating': ratings,
    'sim': sims,
    'processor': processor,
    'ram': ram,
    'battery': battery,
    'display': display,
    'camera': camera,
    'card': card,
    'os': os
})
```
