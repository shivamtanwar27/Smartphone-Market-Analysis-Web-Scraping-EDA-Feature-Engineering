```python
# converting cores values to numeric:

def numeric_cores(core):
    if isinstance(core,str):
        if "Octa" in core:
            return 8
        elif "Hexa" in core:
            return 6
        elif "Deca" in core:
            return 10
        elif "Nine" in core:
            return 9
        elif "Quad" in core:
            return 4
        else:
            return np.nan
    else:
        return np.nan
```

```python
df["processor_cores"] = df.processor_cores.apply(numeric_cores)
```

```python
# correcting the issue in resolution column

df["resolution"] = df.resolution.apply(lambda x : " ".join(x.split()[0:3]).strip() if "Display" in str(x) else x)
```

```python
# creating "resolution_width" column from "resolution"
def extracting_width(resolution):
    if isinstance(resolution,str):
        res = resolution.split("x")
        num1 = int(res[0].strip())
        num2 = int(res[1].strip())
        if num1 < num2:
            return num1
        else:
            return num2
    else:
        return np.nan

df.insert(23,"resolution_width",df.resolution.apply(extracting_width) ) 
```

```python
# creating "resolution_height" column from "resolution"
def extracting_height(resolution):
    if isinstance(resolution,str):
        res = resolution.split("x")
        num1 = int(res[0].strip())
        num2 = int(res[1].strip())
        if num1 > num2:
            return num1
        else:
            return num2
    else:
        return np.nan

df.insert(24,"resolution_height",df.resolution.apply(extracting_height))   
```

```python
# inserting "rear_camera_count" column:

def rear_counter(camera):
    rear_info= camera.split("&")[0].strip()
    counter = rear_info.count("+") + 1
    return counter

df.insert(27, "rear_camera_count",df.camera.apply(rear_counter))
```

```python
# inserting "front_camera_count" column:

def front_counter(camera):

    if len(camera.split("&")) > 1:
        front_info= camera.split("&")[1].strip()
        counter = front_info.count("+") + 1
        return counter
    else:
        return np.nan


df.insert(28, "front_camera_count", df.camera.apply(front_counter))   
```

```python
# inserting "rear_primary_mp" column:

df.insert(29,"rear_primary_mp", df.camera.apply(lambda x : float(x.split()[0].strip())))  
```

```python
# inserting "front_primary_mp" column:

def extracting_front_primary_mp(camera):
    if "&" in camera:
        res = (camera.split("&")[1].strip().split()[0].strip())
        if res == "Main":
            return np.nan
        else:
            return float(res)
    else:
        return np.nan

df.insert(30,"front_primary_mp", df.camera.apply(extracting_front_primary_mp))   
```

```python
# inserting "extended_memory_available" column:

def is_memory_extendable(card):
    if isinstance(card, str):
        if "Not" in card:
            return 0
        else:
            return 1
    else:
        return np.nan


df.insert(32, "extended_memory_available", df.card.apply(is_memory_extendable))  
```

```python
# inserting "extended upto" column:

def extended_limit(card):
    if isinstance(card, str) and "upto" in card:
        res = float(card.split()[-2].strip())

        if res < 3:
            return res*1024
        else: 
            return res
    else:
        return np.nan


df.insert(33,"extended_upto",  df.card.apply(extended_limit)) 
```
